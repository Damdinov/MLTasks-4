```python
from skimage import color
imageGray = color.rgb2gray(image)
mplt.pyplot.imsave('imageGray.png',imageGray)


imageMean = image.copy()
for row in imageMean:
     for column in row:
            mean = np.mean(column[:-1])
            column[0] = mean
            column[1] = mean
            column[2] = mean
mplt.pyplot.imsave('imageMean.png',imageMean)
```

Изображение, полученное с помощью rgb2gray выглядит более качественным, оно и понятно, ведь в rgb2gray используются специально калиброванные веса цветов:
Y = 0.2125 R + 0.7154 G + 0.0721 B
